# Vehicle License Plate Detection and Recognition

Welcome to the project!

In this project, we present a **Vehicle License Plate Detection and Recognition System** designed to efficiently detect and read vehicle license plates. The system supports various input formats, including static images, live video streams, and pre-recorded video files.

The key features of this system include:
- **Visual Output**: Displaying detected license plates directly on the input medium (e.g., video or image).
- **Data Storage**: Saving detected license plate data into a database for further analysis or record-keeping.

This system is suitable for real-world applications such as traffic monitoring, law enforcement, and automated toll systems. The implementation consists of two main components:

1. **License Plate Detection**: Using the **YOLOv10 model**, a state-of-the-art object detection algorithm, to identify and localize license plates within an image or video frame.
2. **Text Recognition**: Employing **Optical Character Recognition (OCR)** techniques to extract and interpret the text from the detected license plate regions.

By combining these two components, we create a robust pipeline capable of detecting, recognizing, and storing license plate information. Additional details about the database integration and extended functionalities are covered in other files of this project.

```python
# Function to analyze dataset

def analyze_yolo_dataset(img_dir, lbl_dir):
    """
    Analyze a YOLO dataset directory and print the analysis results.

    Args:
        img_dir (str): Path to the directory containing images.
        lbl_dir (str): Path to the directory containing label files.
    """
    # Get image and label file lists
    img_files = [f for f in os.listdir(img_dir) if f.endswith(('.jpg', '.png'))]
    lbl_files = [f for f in os.listdir(lbl_dir) if f.endswith('.txt')]

    # Match images and labels
    matched_files = [
        img_file for img_file in img_files
        if os.path.exists(os.path.join(lbl_dir, os.path.splitext(img_file)[0] + '.txt'))]

    unmatched_imgs = set(img_files) - set(matched_files)
    unmatched_lbls = set(lbl_files) - {os.path.splitext(img)[0] + '.txt' for img in matched_files}

    # Analyze label files
    label_counts = Counter()
    total_boxes = 0
    for lbl_file in matched_files:
        lbl_path = os.path.join(lbl_dir, os.path.splitext(lbl_file)[0] + '.txt')
        with open(lbl_path, 'r') as file:
            lines = file.readlines()
            total_boxes += len(lines)
            label_counts.update([int(line.split()[0]) for line in lines if line.strip()])

    # Print results
    print(f"Analysis for directory: {img_dir}")
    print(f"Number of Images: {len(img_files)}")
    print(f"Number of Labels: {len(lbl_files)}")
    print(f"Matched Images: {len(matched_files)}")
    print(f"Unmatched Images: {len(unmatched_imgs)}")
    print(f"Unmatched Labels: {len(unmatched_lbls)}")
    print(f"Total Bounding Boxes: {total_boxes}")
    print(f"Label Distribution: {dict(label_counts)}")
    print("-" * 50)



print("Train Set Analysis:")
analyze_yolo_dataset(img_dir=os.path.join(dataset_base, 'train/images'), lbl_dir=os.path.join(dataset_base, 'train/labels'))

print("\nTest Set Analysis:")
analyze_yolo_dataset(img_dir=os.path.join(dataset_base, 'test/images'), lbl_dir=os.path.join(dataset_base, 'test/labels'))

print("\nValidation Set Analysis:")
analyze_yolo_dataset(img_dir=os.path.join(dataset_base, 'valid/images'), lbl_dir=os.path.join(dataset_base, 'valid/labels'))
```

```python
# Path to dataset images and labels
image_dir = '/content/drive/MyDrive/YOLO/Source_Dataset/train/images'
label_dir = '/content/drive/MyDrive/YOLO/Source_Dataset/train/labels'

# Debug: Check directories
if not os.path.exists(image_dir) or not os.path.exists(label_dir):
    print("One or both directories do not exist!")
else:
    image_files = [img for img in os.listdir(image_dir) if img.endswith(('.jpg', '.png'))]
    if not image_files:
        print("No .jpg or .png files found in the images directory!")
    else:
        print(f"Found {len(image_files)} images.")

# Load 25 images
image_paths = [os.path.join(image_dir, img) for img in image_files[3000:3012]]

# Define a function to parse YOLO format labels
def parse_yolo_labels(label_path, img_shape):
    boxes = []
    with open(label_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            parts = line.strip().split()
            class_id = int(parts[0])
            x_center, y_center, width, height = map(float, parts[1:])
            # Convert YOLO normalized format to absolute pixel coordinates
            img_h, img_w = img_shape[:2]
            x_min = int((x_center - width / 2) * img_w)
            y_min = int((y_center - height / 2) * img_h)
            x_max = int((x_center + width / 2) * img_w)
            y_max = int((y_center + height / 2) * img_h)
            boxes.append((class_id, x_min, y_min, x_max, y_max))
    return boxes

# Create a 5x5 subplot
fig, axes = plt.subplots(3, 4, figsize=(10, 10))
fig.suptitle("Sample Images with Bounding Boxes", fontsize=16)

for i, ax in enumerate(axes.flat):
    if i < len(image_paths):
        img_path = image_paths[i]
        label_path = os.path.join(label_dir, os.path.splitext(os.path.basename(img_path))[0] + '.txt')

        img = cv2.imread(img_path)
        if img is not None:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            if os.path.exists(label_path):
                boxes = parse_yolo_labels(label_path, img.shape)
                for box in boxes:
                    class_id, x_min, y_min, x_max, y_max = box
                    # Draw bounding box
                    cv2.rectangle(img, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
                    # Add class label
                    cv2.putText(img, str(class_id), (x_min, y_min - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
            ax.imshow(img)
            ax.set_title(f"Image {i+1}", fontsize=10)
        else:
            ax.set_title("Failed to load", fontsize=10)
    ax.axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

The dataset appears to be well-labeled, ensuring consistent alignment between images and annotations. However, there are some cases where improvements could have been made to enhance the labeling quality.

### Key Highlights:
- **Diversity**: The dataset is highly diverse, featuring various types of vehicles captured from multiple angles. This diversity makes it representative of real-world scenarios and suitable for practical applications.
- **Scalability Concern**: Despite its strengths, the dataset's size poses a challenge due to limited computational resources.

### Approach:
To address the resource constraints, we plan to utilize only **30%** of the dataset from each folder (Train, Test, and Validation). This will allow us to create a **custom subset** that retains the diversity and quality of the original dataset while remaining manageable for our available computation power.

```python
# Paths to original dataset directories (relative paths)
input_dataset_path = '/content/drive/MyDrive/YOLO/Source_Dataset'
train_img_dir = os.path.join(input_dataset_path, 'train/images')
train_lbl_dir = os.path.join(input_dataset_path, 'train/labels')
valid_img_dir = os.path.join(input_dataset_path, 'valid/images')
valid_lbl_dir = os.path.join(input_dataset_path, 'valid/labels')
test_img_dir = os.path.join(input_dataset_path, 'test/images')
test_lbl_dir = os.path.join(input_dataset_path, 'test/labels')

# Paths to subset directories (relative to the working directory)
subset_dir = 'yolo-custom-dataset'
subset_train_img_dir = os.path.join(subset_dir, 'train/images')
subset_train_lbl_dir = os.path.join(subset_dir, 'train/labels')
subset_valid_img_dir = os.path.join(subset_dir, 'valid/images')
subset_valid_lbl_dir = os.path.join(subset_dir, 'valid/labels')
subset_test_img_dir = os.path.join(subset_dir, 'test/images')
subset_test_lbl_dir = os.path.join(subset_dir, 'test/labels')

# Ensure subset directories exist
os.makedirs(subset_train_img_dir, exist_ok=True)
os.makedirs(subset_train_lbl_dir, exist_ok=True)
os.makedirs(subset_valid_img_dir, exist_ok=True)
os.makedirs(subset_valid_lbl_dir, exist_ok=True)
os.makedirs(subset_test_img_dir, exist_ok=True)
os.makedirs(subset_test_lbl_dir, exist_ok=True)

def create_subset(img_dir, lbl_dir, subset_img_dir, subset_lbl_dir, fraction=0.25):
    """
    Create a subset of the dataset with adjusted labels for one class.
    Args:
        img_dir (str): Directory containing image files.
        lbl_dir (str): Directory containing label files.
        subset_img_dir (str): Destination directory for subset images.
        subset_lbl_dir (str): Destination directory for subset labels.
        fraction (float): Fraction of the dataset to include in the subset.
    """
    # List all image files
    img_files = [f for f in os.listdir(img_dir) if f.endswith(('.jpg', '.png'))]

    # Select a fraction of files randomly
    subset_size = max(1, int(len(img_files) * fraction))  # Ensure at least 1 file
    subset_files = random.sample(img_files, subset_size)

    for img_file in subset_files:
        # Copy image file
        shutil.copy(os.path.join(img_dir, img_file), os.path.join(subset_img_dir, img_file))

        # Copy corresponding label file and adjust for one class
        lbl_file = os.path.splitext(img_file)[0] + '.txt'
        lbl_src_path = os.path.join(lbl_dir, lbl_file)
        lbl_dest_path = os.path.join(subset_lbl_dir, lbl_file)
        if os.path.exists(lbl_src_path):
            with open(lbl_src_path, 'r') as src_file, open(lbl_dest_path, 'w') as dest_file:
                for line in src_file:
                    # Replace all labels with class 0 (one class)
                    parts = line.split()
                    parts[0] = '0'  # Set class to 0
                    dest_file.write(' '.join(parts) + '\n')

# Create subsets
create_subset(train_img_dir, train_lbl_dir, subset_train_img_dir, subset_train_lbl_dir, fraction=0.3)
create_subset(valid_img_dir, valid_lbl_dir, subset_valid_img_dir, subset_valid_lbl_dir, fraction=0.3)
create_subset(test_img_dir, test_lbl_dir, subset_test_img_dir, subset_test_lbl_dir, fraction=0.3)

# YAML file for the subset dataset with one class
yaml_data = {
    'input_dataset_path': input_dataset_path,
    'train': '../train/images',
    'val': '../valid/images',
    'test': '../test/images',
    'nc': 1,  # Number of classes (one class)
    'names': ['License_Plate']  # Name for the single class
}

# Save the new YAML file
yaml_path = os.path.join(subset_dir, 'dataset.yaml')
with open(yaml_path, 'w') as yaml_file:
    yaml.dump(yaml_data, yaml_file, default_flow_style=False)

print(f"Subset dataset created at: {subset_dir}")
print(f"YAML file saved to: {yaml_path}")
```
